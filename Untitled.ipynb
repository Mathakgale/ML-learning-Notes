# Linear Regression

Is a supervised ML model. 

ML model has three processes.
- prediction model
- cost function
- optimizer 

## 1. prediction model

$$\hat{y}=\theta_{0}+\theta_{1}x_{1}+\theta_{2}x_{2}+...+\theta_{n}x_{n}$$

where: 

$$\hat{y}$$ - prediction value 

$$ n $$- is the number of features 

$$ x_{i}$$ is the ith feature value

$$ \theta_{j}$$ is the jth model parameter

## 2. Cost function

- used to minimize error so that the actual value == predicted value
- optimize weights
- for linear regrassion we have cost function called the Mean Square Error (**MSE**)

$$MSE =  \frac{1}{m} \sum_{m}^{i=1}(\theta^{T}x^{(i)} -y^{(i)})^{2}$$

## 3. Optimization function

- derive the cost with each theta

$$\frac{\partial \theta}{\partial x} MSE(\theta) $$


Gradient descent

$$\theta^{next step}=\theta- \eta \triangledown MSE(\theta)$$
